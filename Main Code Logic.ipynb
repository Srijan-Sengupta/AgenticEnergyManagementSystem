```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_ollama import ChatOllama

llm = ChatOllama(
    model="deepseek-r1:7b",
    temperature=0.0,
    reasoning=True
)

template = """Question: {question}

Answer: Let's think step by step."""

prompt = ChatPromptTemplate.from_template(template)

chain = prompt | llm
val = chain.invoke({"question": "Using Java swing code a complete runnable hermite spline that lets you draw using 4 clicks"})
```

```python
print(val.additional_kwargs.get("reasoning_content"))

print("\n\n\nThinking Done\n\n\n")

print(val.content)
```
